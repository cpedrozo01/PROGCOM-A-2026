```python
numero = int(input("Ingresa un número: "))

print("Usando ciclo for:")
for i in range(1, 11):
    print(numero, "x", i, "=", numero * i)

print("\nUsando ciclo while:")
i = 1
while i <= 10:
    print(numero, "x", i, "=", numero * i)
    i += 1
```

```python
n = int(input("Ingresa un número n: "))

# Con for
suma = 0
for i in range(1, n + 1):
    suma += i
print("Suma con for:", suma)

# Con while
suma = 0
i = 1
while i <= n:
    suma += i
    i += 1
print("Suma con while:", suma)
```

```python
num = int(input("Ingresa un número: "))

if num < 0:
    print("No existen factoriales negativos")
else:
    # Con for
    factorial = 1
    for i in range(1, num + 1):
        factorial *= i
    print("Factorial con for:", factorial)

    # Con while
    factorial = 1
    i = 1
    while i <= num:
        factorial *= i
        i += 1
    print("Factorial con while:", factorial)
```

```python
n = int(input("¿Cuántos términos deseas?: "))

# Con for
a, b = 0, 1
print("Fibonacci con for:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b

# Con while
print("\nFibonacci con while:")
a, b = 0, 1
i = 0
while i < n:
    print(a, end=" ")
    a, b = b, a + b
    i += 1
```

```python
colores = []
print("Define 10 colores:")
for i in range(10):
    color = input(f"Color {i+1}: ")
    colores.append(color)

secreto = colores[0]  # Puedes cambiar la lógica si deseas
intentos = 0
adivinado = False

while intentos < 10 and not adivinado:
    intento = input("Adivina el color: ")
    intentos += 1

    if intento == secreto:
        print("¡Correcto!")
        adivinado = True
    else:
        print("Incorrecto, intenta nuevamente.")

if not adivinado:
    print("Se acabaron los intentos.")
```

```python
# Con for
print("Pares con for:")
for i in range(1, 101):
    if i % 2 == 0:
        print(i, end=" ")

# Con while
print("\nPares con while:")
i = 1
while i <= 100:
    if i % 2 == 0:
        print(i, end=" ")
    i += 1
```

```python
num = int(input("Ingresa un número: "))
numero = abs(num)

# Con while
suma = 0
while numero > 0:
    suma += numero % 10
    numero //= 10

print("Suma de dígitos:", suma)
```

```python
num = int(input("Ingresa un número: "))

if num <= 1:
    print("No es primo")
else:
    es_primo = True
    for i in range(2, num):
        if num % i == 0:
            es_primo = False
            break

    if es_primo:
        print("Es primo")
    else:
        print("No es primo")
```

```python
filas = int(input("Número de filas: "))

# Con for
print("Triángulo con for:")
for i in range(1, filas + 1):
    print("*" * i)

# Con while
print("Triángulo con while:")
i = 1
while i <= filas:
    print("*" * i)
    i += 1
```

```python
base = int(input("Base: "))
exponente = int(input("Exponente: "))

resultado = 1

# Con for
for i in range(exponente):
    resultado *= base

print("Potencia con for:", resultado)

# Con while
resultado = 1
i = 0
while i < exponente:
    resultado *= base
    i += 1

print("Potencia con while:", resultado)
```
