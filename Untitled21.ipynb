```python
#EJERCICIO 5
estudiantes = {
    "Jp": 85,
    "Simon": 92,
    "Cristian": 78,
    "Zharick": 95,
    "Sebastian ": 88
}

def promedio_todos():
    total = 0
    for nota in estudiantes.values():
        total = total + nota
    promedio = total / len(estudiantes)
    print("El promedio de todos los estudiantes es:", promedio)

def promedio_estudiante(nombre):
    if nombre in estudiantes:
        print("La nota de", nombre, "es:", estudiantes[nombre])
    else:
        print("El estudiante no existe en el diccionario.")

promedio_todos()
promedio_estudiante("Zharick")
```

```python
#EJERCICIO 6
personajes_rpg = {
    "guerrero": {"nivel": 15, "arma": "espada"},
    "mago": {"nivel": 12, "arma": "bastón"},
    "arquero": {"nivel": 10, "arma": "arco"}
}

clave = "hechicero"

if clave in personajes_rpg:
    print(f" '{clave}' encontrado:", personajes_rpg[clave])
else:
    print(f" Advertencia: La clave '{clave}' no existe en el diccionario.")
```

```python
#EJERCICIO 7
palabra = "magia"
conteo = {}

for letra in palabra:
    if letra in conteo:
        conteo[letra] = conteo[letra] + 1
    else:
        conteo[letra] = 1

print("Conteo de letras en", palabra, ":")
for letra in conteo:
    print(letra, ":", conteo[letra])
```

```python
#EJERCICIO 8
elementos = ["gema", "oro", "gema", "madera", "oro", "oro", "gema"]
conteo = {}

for elemento in elementos:
    if elemento in conteo:
        conteo[elemento] = conteo[elemento] + 1
    else:
        conteo[elemento] = 1

print("Elementos recolectados en la expedición:")
for elemento in conteo:
    print(elemento, ":", conteo[elemento])
```

```python
#EJERCICIO 10
inventario = {
    "manzana": 10,
    "pan": 4,
    "leche": 6
}

print("Inventario inicial:", inventario)


inventario["huevos"] = 12
print("\nDespués de agregar 'huevos':", inventario)


inventario["leche"] = 9
print("Después de actualizar 'leche' a 9:", inventario)

del inventario["pan"]
print("Después de eliminar 'pan':", inventario)

print("\nInventario final:")
for objeto in inventario:
    print(" -", objeto, ":", inventario[objeto])
```
