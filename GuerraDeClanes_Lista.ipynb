```python
 #lista de trabajadores
 n = int(input("Ingrese la cantidad de trabajadores: "))

sueldos = []
nombres = []

for i in range(n):
    nombre = input("Nombre del trabajador: ")
    sueldo = float(input("Sueldo: "))

    nombres.append(nombre)
    sueldos.append(sueldo)

promedio = sum(sueldos) / n
print("Promedio de sueldos:", promedio)

contador = 0

for i in range(n):
    if sueldos[i] > promedio:
        contador += 1
        print("Gana más que el promedio normal:", nombres[i])

print("Cantidad que ganan más que el promedio:", contador)
```

```python
#Menú de tiempo
import random

productos = ["Pan", "Leche", "Huevos", "Arroz"]
precios = [2000, 3500, 5000, 4200]

while True:
    print("\n--- MENÚ TIENDA ---")
    print("1. Ver productos")
    print("2. Agregar producto")
    print("3. Producto en oferta")
    print("4. Salir")

    opcion = int(input("Elija una opción: "))

    if opcion == 1:
        print("\nLista de productos:")
        for i in range(len(productos)):
            print(productos[i], "- $", precios[i])

    elif opcion == 2:
        nuevo = input("Ingrese el nombre del producto: ")
        precio = float(input("Ingrese el precio: "))

        productos.append(nuevo)
        precios.append(precio)

        print("Producto agregado")

    elif opcion == 3:
        indice = random.randint(0, len(productos)-1)
        print("\nProducto en oferta hoy:")
        print(productos[indice], "- $", precios[indice])

    elif opcion == 4:
        print("Fin del programa")
        break

    else:
        print("Debe elegir una opción válida")
```

```python
# Lista de números
numeros = [10, 15, 20, 25, 30, 33, 40]

# Número n
n = int(input("Ingrese el número n: "))

contador = 0

for num in numeros:
    if num % n == 0:
        contador += 1

print("Cantidad de múltiplos de", n, ":", contador)
```

```python
# Lista de pacientes
pacientes = ["Ana", "Luis", "Carlos", "Maria", "Pedro"]

# Lista de fechas de vacunación (misma posición que el paciente)
fechas_vacunacion = [
    "10/01/2024",
    "15/01/2024",
    "20/01/2024",
    "25/01/2024",
    "30/01/2024"
]

# Cantidad de pacientes vacunados
cantidad = len(pacientes)

print("Cantidad de pacientes vacunados:", cantidad)

print("Pacientes y fechas de vacunación:")

for i in range(cantidad):
    print(pacientes[i], "- Fecha:", fechas_vacunacion[i])
```
