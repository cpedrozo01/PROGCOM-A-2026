```python
# 1. TABLA DE MULTIPLICAR
num_tabla = int(input("1. ¿De qué número quieres la tabla?: "))
contador = 1
while contador <= 10:
    resultado = num_tabla * contador
    print(num_tabla, "x", contador, "=", resultado)
    contador = contador + 1
```

```python
2. SUMA DE NATURALES
limite_suma = int(input("\n2. ¿Hasta qué número quieres sumar?: "))
suma_total = 0
contador = 1
while contador <= limite_suma:
    suma_total = suma_total + contador
    contador = contador + 1
print("La suma de todos es:", suma_total)
```

```python
# 3. FACTORIAL
n_fact = int(input("\n3. Dime un número para el factorial: "))
if n_fact < 0:
    print("No existen factoriales de negativos.")
else:
    resultado_fact = 1
    contador = 1
    while contador <= n_fact:
        resultado_fact = resultado_fact * contador
        contador = contador + 1
    print("El factorial es:", resultado_fact)
```

```python
# 4. SERIE FIBONACCI
n_fibo = int(input("\n4. ¿Cuántos números de Fibonacci quieres?: "))
a = 0
b = 1
contador = 0
while contador < n_fibo:
    print(a, end=" ")
    auxiliar = a + b
    a = b
    b = auxiliar
    contador = contador + 1
print() # Salto de línea
```

```python
# 6. NÚMEROS PARES (1 al 100)
print("\n6. Mostrando números pares:")
contador = 2
while contador <= 100:
    print(contador, end=" ")
    contador = contador + 2
```

```python
# 7. SUMA DE DÍGITOS
num_digitos = int(input("\n\n7. Escribe un número largo: "))
suma_d = 0
# Usamos un truco matemático: dividir entre 10 para quitar el último número
while num_digitos > 0:
    ultimo = num_digitos % 10
    suma_d = suma_d + ultimo
    num_digitos = num_digitos // 10 # El // es para división entera
print("La suma de sus dígitos es:", suma_d)
```

```python
# 8. ES PRIMO
n_primo = int(input("\n8. ¿Qué número quieres revisar si es primo?: "))
divisor = 2
es_primo = True
if n_primo < 2:
    es_primo = False
while divisor < n_primo:
    if n_primo % divisor == 0:
        es_primo = False
    divisor = divisor + 1
print("¿Es primo?:", es_primo)
```

```python
# 9. TRIÁNGULO DE ASTERISCOS
pisos = int(input("\n9. ¿De cuántos pisos quieres el triángulo?: "))
fila_actual = 1
while fila_actual <= pisos:
    # En Python podemos multiplicar texto por números
    print("*" * fila_actual)
    fila_actual = fila_actual + 1
```

```python
# 10. POTENCIA SIN OPERADOR **
base = int(input("\n10. Dime la base: "))
exponente = int(input("Dime el exponente: "))
potencia_total = 1
contador = 1
while contador <= exponente:
    potencia_total = potencia_total * base
    contador = contador + 1
print("El resultado es:", potencia_total)
```
