```python
nombre_producto = "Silla ergonomica"
inventario = 20
stock_minimo = 5
demanda = [15, 20, 18, 25, 22, 30]

print("=" * 50)
print("   PLAN DE PRODUCCION:", nombre_producto)
print("=" * 50)
print(f"{'Semana':<10} {'Demanda':<10} {'Inventario':<12} {'Producir':<10}")
print("-" * 50)

for i in range(len(demanda)):
    semana = i + 1
    pedido = demanda[i]
    if inventario < pedido + stock_minimo:
        producir = pedido + stock_minimo - inventario
    else:
        producir = 0
    inventario = inventario + producir - pedido
    print(f"{semana:<10} {pedido:<10} {inventario:<12} {producir:<10}")

print("=" * 50)
print("Fin del plan de produccion")
print()
print("=" * 50)
print("   MATERIALES NECESARIOS (por unidad producida)")
print("=" * 50)

materiales = [
    ["Estructura metalica", 1],
    ["Asiento tapizado",    1],
    ["Ruedas",              5],
    ["Tornillos M6",        8],
]
inventario2 = 20
total_producido = 0

for i in range(len(demanda)):
    pedido = demanda[i]
    if inventario2 < pedido + stock_minimo:
        producir = pedido + stock_minimo - inventario2
    else:
        producir = 0
    total_producido = total_producido + producir
    inventario2 = inventario2 + producir - pedido

print(f"Total a producir: {total_producido} unidades")
print()

for material in materiales:
    nombre_mat = material[0]
    cantidad   = material[1]
    total_mat  = cantidad * total_producido
    print(f"  {nombre_mat}: {total_mat} unidades")

print("=" * 50)
```
